### Finding missing values which can be dropped or imputed

* First we find the values which are missing and sort them in descending order

**FOR PREVIOUS APPLICANTS DATAFRAME:**

```python
# Getting all the columns which have null or na in descending order - previous applicants data
rows = previous_app_df.shape[0]
missing_values_prev = previous_app_df.isnull().sum().sort_values(ascending=False).apply(lambda x: round(x / rows * 100,2))
# Filtering values which have some missing values
missing_values_prev = missing_values_prev[missing_values_prev >0]
missing_values_prev
```

> In the above previous applicants dataframe, values >= 40% missing can be considered as incomplete data, and i will be excluding them in my analysis

**FOR CURRENT APPLICANTS DATAFRAME:**

```python
# Getting all the columns which have null or na in descending order
rows = current_app_df.shape[0]
missing_values_current = current_app_df.isnull().sum().sort_values(ascending=False).apply(lambda x: round(x / rows * 100,2))
# Filtering values which have some missing values
missing_values_current = missing_values_current[missing_values_current >0]
missing_values_current
```

* Now, lets set a threshhold.Observing the columns to be removed are the ones which half values **>= 47%** missing , and the columns which can be imputed are the ones with values **<= 47** percent missing.

```python
# Removing all percentages >= 47 As 47 percent is a significant chunk of data missing.
cols_to_be_dropped = missing_values_current[missing_values >= 47]
cols_to_be_imputed = missing_values_current[missing_values < 47]

## Converting the same to lists
# List of columns to be imputed
cols_to_be_imputed_list = cols_to_be_imputed.index.tolist()
# List of columns to be dropped
cols_to_be_dropped_list = cols_to_be_dropped.index.tolist()
```

#### **Imputing missing value columns**

* Lets work on the columns that can be fixed with imputations first. The list is:

```python
cols_to_be_imputed_list
```

> 1. **OCCUPATION_TYPE** being a categorical variable, can be imputed with the mode of the variable which turns out to be **Laborers**.

```python
current_app_df['OCCUPATION_TYPE'].describe()
```

```python
# Find mode
occupation_mode = current_app_df['OCCUPATION_TYPE'].mode()[0]
## Setting the null variables to mode
current_app_df['OCCUPATION_TYPE'].fillna(occupation_mode,inplace=True)

## Check that the value is no longer null
current_app_df['OCCUPATION_TYPE'].isnull().sum()
```

> 2. Feature **'EXT_SOURCE_2'** is a continuous variable, here we can replace the values that are missing with the median value.

```python
# Checking the variable
current_app_df['EXT_SOURCE_2'].describe()
# Calculating the median
median_ext_score_2 = current_app_df['EXT_SOURCE_2'].median()
```

```python
# Replacing the missing value with the median value
current_app_df['EXT_SOURCE_2'].fillna(median_ext_score_2,inplace=True)
# Confirming that the value is no longer null
current_app_df['EXT_SOURCE_2'].isnull().sum()
```

> 3. Similarly, Feature **'EXT_SOURCE_3'** is a continuous variable, here we can replace the values that are missing with the median value.

```python
# Checking the variable
current_app_df['EXT_SOURCE_3'].describe()
# Calculating the median
median_ext_score_3 = current_app_df['EXT_SOURCE_3'].median()
```

```python
# Replacing the missing value with the median value
current_app_df['EXT_SOURCE_3'].fillna(median_ext_score_3,inplace=True)
# Confirming that the value is no longer null
current_app_df['EXT_SOURCE_3'].isnull().sum()
```

> 4. Feature **'NAME_TYPE_SUITE'**  is a categorical variable which means who accompanied the loan applicant when he was applying for the loan. Here, again we can **impute the value with mode**

```python
# Understanding the variable
current_app_df['NAME_TYPE_SUITE'].value_counts()
# Calculating mode
name_type_mode = current_app_df['NAME_TYPE_SUITE'].mode()[0]
```

```python
# Replacing the variable with mode
current_app_df['NAME_TYPE_SUITE'].fillna(name_type_mode,inplace=True)
# Checking values if null
current_app_df['NAME_TYPE_SUITE'].isnull().sum()
```

> 5. Feature **'AMT_GOODS_PRICE'**  is a continuous variable, which indicates the price of amount of goods **IF APPLICABLE**. Here **it doesn't make sense to replace the value with median or mean** as the applicant might not have a price of amount of goods in his loan application. Here , we can either 

* Drop the rows with such values - for sake of analysis
* Impute the rows with value 0

I will be imputing the rows with 0 wherever the data is missing.

```python
# Understanding the variable - 278 values are missing
current_app_df['AMT_GOODS_PRICE'].isnull().sum()
```

```python
# Seeing the range of the values
current_app_df['AMT_GOODS_PRICE'].describe()
```

```python
# All the missing value are NaN
current_app_df['AMT_GOODS_PRICE'][current_app_df['AMT_GOODS_PRICE'].isnull()]
current_app_df['AMT_GOODS_PRICE'].fillna(0,inplace=True)
```

```python
# Checking the values for nulls after replacement
current_app_df['AMT_GOODS_PRICE'].isnull().sum()
```

### Cleaning and manipulating variables for better analysis

* Some variables in our dataset, for eg **DAYS_BIRTH** , **DAYS_EMPLOYED** are negative, and we can convert these in years so it helps in our analysis.
 
 1. For **DAYS_BIRTH** variable we introduce a new variable called **AGE_YEARS**(In years) and **AGE_RANGE**(range of years)


 

```python
# Since age is in days, we should convert it to years, Also Age is negative, we should take the absolute value of the same.
current_app_df['AGE_YEARS'] =  -current_app_df['DAYS_BIRTH'] // 365
current_app_df['AGE_YEARS'].describe()
```

```python
## Minimum age is 20, lets categories the age into age groups based on 5 years STEP
min_age=20
STEP = 5
bins = [min_age + i * STEP for i in range(11)]
categories=[str(x) +'-'+ str(x+STEP) for x in bins[:-1]]

# With bins and categories, define a new variable age range
current_app_df['AGE_RANGE'] = pd.cut(current_app_df['AGE_YEARS'],bins,labels=categories)

```

```python
# Seeing the distribution of the same
current_app_df['AGE_RANGE'].value_counts(normalize=True)
```

```python
# Converting to integer
current_app_df['AMT_INCOME_TOTAL_INT'] = current_app_df['AMT_INCOME_TOTAL'].astype(int)
# Understanding the range of data and where most of the data lies
current_app_df['AMT_INCOME_TOTAL_INT'].describe()

# Our data lies in this range , lets define bins and categories for the same
bins = [25000,75000,125000,175000,225000,275000,325000,375000,np.inf]
categories = ['25K-75K','75K-125K','125K-175K','175K-225K','225K-275K','275K-325K','325K-375k','375K+']
# Introducing a new variable income_range which will help analyse ranges of income where people fall in.
current_app_df['INCOME_RANGE'] = pd.cut(current_app_df['AMT_INCOME_TOTAL_INT'],bins,labels=categories)

current_app_df['INCOME_RANGE'].value_counts(normalize=True)
```

 3. Finally , For **DAYS_EMPLOYED** variable we introduce a new variable called **YEARS_EMPLOYED**(In years) and **EMPLOYMENT_RANGE**(range of years)

```python
current_app_df['DAYS_EMPLOYED'].describe()
```

```python
# Adding negative sign for employed years
current_app_df['YEARS_EMPLOYED'] = -round(current_app_df['DAYS_EMPLOYED'] / 365,1)
# Getting a basic description of new column
current_app_df['YEARS_EMPLOYED'].describe()
```

```python
# Checking minimum value without considering the new negative years
current_app_df.YEARS_EMPLOYED[current_app_df['YEARS_EMPLOYED'] >= 0].describe()
## The output implies that some of the people with 0 years of employment have never been employed.
```

> It seems we have a value of **365243.000000** throughout our dataset for **DAYS_EMPLOYED** feature, which could signify a special value. I am assuming this value to mean **DATA NOT AVAILABLE** , and replacing the same in a new column with NA. Also a very few amount of values have employed days as 0. These could be the people who are unemployed. I am combinining both of them in a special bin **'NA/UNEMPLOYED'**. I am also introducing a new range variable called **YEARS_EMPLOYED_RANGE** which will help with visualizations and analysis.

```python
# Number of special values
current_app_df['DAYS_EMPLOYED'][current_app_df['DAYS_EMPLOYED'] == 365243.000000].count()
```

```python
# As we still have negative values we can have to replace the same using binning specially
min_years=0
# Binning in steps of 3
STEP = 3
bins = [min_years + i * STEP for i in range(7)]

# Defining steps and categories
categories=[str(x) +'-'+ str(x+STEP) for x in bins[:-1]]
bins = [-np.inf] + bins + [np.inf]
categories =  ['NA/UNEMPLOYED'] + categories + ['18+']

# With bins and categories, define a new variable YEARS_EMPLOYED_RANGE 
current_app_df['YEARS_EMPLOYED_RANGE'] = pd.cut(current_app_df['YEARS_EMPLOYED'],bins,labels=categories)
# Seeing the results
current_app_df['YEARS_EMPLOYED_RANGE'].value_counts()
```
