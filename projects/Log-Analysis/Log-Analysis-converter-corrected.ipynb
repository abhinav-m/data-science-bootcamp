```python

__VM_IP__ = "172.24.88.49"
ERROR_CHARS = "ERROR AT : "
__PLATFORM_CHARS__ = "PLATFORM:"
__REQ_URL_CHARS__ = "Requested URL :"
__REQ_FEED_CHARS__ = "Requested Feed :"
__ERR_DESC_CHARS = "Error:"
COMMA = ","

PROD_URL  = "https://navbharattimes.indiatimes.com"
def remove_characters_and_comma(string,characters):
    new_string =  string.replace(characters,"").strip()
    return new_string.replace(",","")


# output_csvfile = open('lang-error-17-12-2020.csv',"w+")



# counter = 1
# TIME_LOG_LINE = reader.readline()
#     ## First line is date
# print(remove_characters_and_comma(TIME_LOG_LINE,ERROR_CHARS))

files_to_convert = ["lang-error-21-12-2020","lang-error-22-12-2020","lang-error-23-12-2020","lang-error-26-12-2020","lang-error-27-12-2020"]

combined_output_csv = open("combined_error_log.csv","w+")
combined_output_csv.write("IP,TIME,PLATFORM,REQ_URL,REQ_FEED,DESC\n")

for file_name in files_to_convert:
    with open(file_name+".log","r") as reader:
        output_csvfile = open(file_name +".csv","w+")
        output_csvfile.write("IP,TIME,PLATFORM,REQ_URL,REQ_FEED,DESC\n")


        def parse_and_append_to_csv(line,line_num):
            if line_num == 1:
                output_csvfile.write(__VM_IP__+",")
                combined_output_csv.write(__VM_IP__+",")
                parsed_output = remove_characters_and_comma(line,ERROR_CHARS) +","
                combined_output_csv.write(parsed_output)
                output_csvfile.write(parsed_output)
            elif line_num ==2:
                parsed_output = remove_characters_and_comma(line,__PLATFORM_CHARS__) +","
                output_csvfile.write(parsed_output)
                combined_output_csv.write(parsed_output)
            elif line_num == 3:
                parsed_output = remove_characters_and_comma(PROD_URL + line,__REQ_URL_CHARS__) +","
                output_csvfile.write(parsed_output)
                combined_output_csv.write(parsed_output)
            elif line_num == 4:
                if __REQ_FEED_CHARS__ in line:
                    parsed_output = remove_characters_and_comma(line,__REQ_FEED_CHARS__) + ","
                    output_csvfile.write(parsed_output)
                    combined_output_csv.write(parsed_output)
                    error_line = "'"+reader.readline().strip().replace(",","") +"'"+ "\n"
                    output_csvfile.write(error_line)
                    combined_output_csv.write(error_line)
                else:
                    # Feed not available, therefore fill with NA
                    output_csvfile.write('NA,')
                    combined_output_csv.write('NA,')
                    parsed_output = "'"
                    while  COMMA not in line and line != "":
                        parsed_output += line.strip()
                        line = reader.readline()
                    parsed_output += line.strip().replace(",","")  +"'"  +"\n"
    #                 print(parsed_output)
                    output_csvfile.write(parsed_output)
                    combined_output_csv.write(parsed_output)

        counter = 1
        for line in reader:
            #  Skip new line characters
            if line =="\n":
                continue
            parse_and_append_to_csv(line,counter)
            if counter == 4:
                counter = 1
            else:
                counter+= 1
        output_csvfile.close()

combined_output_csv.close()

# with open('lang-error-17-12-2020.log',"r") as reader:
#     output_csvfile = open('lang-error-17-12-2020-TEST-2.csv',"w+")
#     output_csvfile.write("TIME,PLATFORM,REQ_URL,REQ_FEED,DESC\n")


#     def parse_and_append_to_csv(line,line_num):
#         print(line,line_num)
#         if line_num == 1:
#             parsed_output = remove_characters_and_comma(line,ERROR_CHARS) +","
#             output_csvfile.write(parsed_output)
#         elif line_num ==2:
#             parsed_output = remove_characters_and_comma(line,__PLATFORM_CHARS__) +","
#             output_csvfile.write(parsed_output)
#         elif line_num == 3:
#             parsed_output = remove_characters_and_comma(PROD_URL + line,__REQ_URL_CHARS__) +","
#             output_csvfile.write(parsed_output)
#         elif line_num == 4:
#             if __REQ_FEED_CHARS__ in line:
#                 parsed_output = remove_characters_and_comma(line,__REQ_FEED_CHARS__) + ","
#                 output_csvfile.write(parsed_output)
#                 error_line = reader.readline().strip() + "\n"
#                 output_csvfile.write(error_line)
#             else:
#                 # Feed not available, therefore fill with NA
#                 output_csvfile.write('NA,')
#                 parsed_output = "'"
#                 while  COMMA not in line and line != "":
#                     parsed_output += line.strip()
#                     line = reader.readline()
#                 parsed_output += line.strip().replace(",","")  +"'"  +"\n"
# #                 print(parsed_output)
#                 output_csvfile.write(parsed_output)

#     counter = 1
    
#     for line in reader:
#         #  Skip new line characters
#         if line =="\n":
#             continue
#         # Append ip , might need for merging
#         output_csvfile.write(__VM_IP__+",")
#         parse_and_append_to_csv(line,counter)
#         if counter == 4:
#             counter = 1
#         else:
#             counter+= 1
#     output_csvfile.close()
    

```
