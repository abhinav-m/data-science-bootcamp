```python
# Getting the tv values we want to check
queried_df = title_basics_df.query('titleType == "tvEpisode" ')
queried_df['startYear'] = pd.to_numeric(queried_df['startYear'], errors='coerce')
queried_df = queried_df.sort_values('startYear',ascending=False)
queried_df = queried_df.query('originalTitle == "Secrets" and startYear == 2017 ')
queried_df.head(20)
```
